## 4Ô∏è‚É£ Entropy Awal

Rumus:

```math
Entropy(D) = - \sum p_i \log_2(p_i)

Baik! Sekarang kita akan menghitung ulang perhitungan manual *Information Gain* dan *Gini Index* **untuk fitur Berat Buah (gram)** berdasarkan data terbaru:

---

### 1Ô∏è‚É£ **Urutkan data berdasarkan Berat Buah (gram):**

| No | Berat Buah (gram) | Kelas           |
| -- | ----------------- | --------------- |

| 8  | 1850              | Cucumis melo    |
| 7  | 1900              | Cucumis melo    |
| 10 | 2000              | Cucumis melo    |

---

### 2Ô∏è‚É£ **Ambil threshold split:**

Kita gunakan threshold = **800 gram**
(sebab semua Cucumis sativus < 800 gram, semua Cucumis melo > 800 gram ‚Äî terjadi perpindahan mayoritas kelas)

#### Maka:

* **Group Left (‚â§ 800 gram)** ‚Üí 8 data (semua Cucumis sativus)
* **Group Right (> 800 gram)** ‚Üí 7 data (semua Cucumis melo)

---

### 3Ô∏è‚É£ **Komposisi kelas:**

* Sebelum split:

  * 8 *Cucumis sativus*
  * 7 *Cucumis melo*

$$
Entropy(D) = -\frac{8}{15} \log_2 \frac{8}{15} - \frac{7}{15} \log_2 \frac{7}{15}
$$

Sama seperti sebelumnya:

$$
Entropy(D) \approx 0.994
$$

---

### 4Ô∏è‚É£ **Entropy setelah split:**

#### Group Left (‚â§ 800 gram):

* 8 *Cucumis sativus*, 0 *Cucumis melo*

$$
Entropy(left) = -\frac{8}{8}\log_2\frac{8}{8} = 0
$$

#### Group Right (> 800 gram):

* 7 *Cucumis melo*, 0 *Cucumis sativus*

$$
Entropy(right) = -\frac{7}{7}\log_2\frac{7}{7} = 0
$$

---

### 5Ô∏è‚É£ **Information Gain:**

$$
IG = Entropy(D) - \left( \frac{8}{15} \cdot 0 + \frac{7}{15} \cdot 0 \right)
$$

$$
IG = 0.994 - 0 = \boxed{0.994}
$$

---

### 6Ô∏è‚É£ **Gini Index**

#### Gini Awal:

$$
Gini(D) = 1 - \left( \frac{8}{15} \right)^2 - \left( \frac{7}{15} \right)^2
$$

$$
Gini(D) \approx 1 - 0.284 - 0.218 \approx 0.498
$$

#### Gini(Left):

Semua satu kelas ‚Üí Gini = 0

#### Gini(Right):

Semua satu kelas ‚Üí Gini = 0

#### Gini Split:

$$
Gini_{split} = \frac{8}{15} \cdot 0 + \frac{7}{15} \cdot 0 = \boxed{0}
$$

---

### 7Ô∏è‚É£ **Hasil Akhir Fitur "Berat Buah (gram)"**

| Ukuran Evaluasi  | Nilai |
| ---------------- | ----- |
| Entropy Awal     | 0.994 |
| Information Gain | 0.994 |
| Gini Split       | 0.000 |

---

### üîé **Kesimpulan**

Fitur Berat Buah (gram) menghasilkan split sempurna, karena seluruh data bisa dipisahkan sempurna dengan threshold 800 gram.

## Visualisasi Decision Tree

Ini adalah code program untuk mencari decision tree sekaligus rules

````python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import LabelEncoder

# === Data 3 Kelas (Sesuai perhitungan kita sebelumnya) ===
data = {
    "Diameter Buah (cm)": [4.2, 4.5, 4.0, 3.8, 4.3,  # sativus
                          16.0, 17.0, 15.0, 17.5, 14.0, # melo
                          7.0, 7.5, 6.8, 8.0, 7.2],   # metuliferus
    "Berat Buah (gram)": [500, 530, 480, 430, 510,    # sativus
                          1800, 1900, 1750, 2000, 1600, # melo
                          900, 950, 870, 1000, 920],  # metuliferus
    "Kelas": [
        "Cucumis sativus", "Cucumis sativus", "Cucumis sativus", "Cucumis sativus", "Cucumis sativus",
        "Cucumis melo", "Cucumis melo", "Cucumis melo", "Cucumis melo", "Cucumis melo",
        "Cucumis metuliferus", "Cucumis metuliferus", "Cucumis metuliferus", "Cucumis metuliferus", "Cucumis metuliferus"
    ]
}


df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])

# Menampilkan hasil encoding agar jelas
print("=== Hasil Label Encoding ===")
for index, nama_kelas in enumerate(le.classes_):
    print(f"{nama_kelas} -> {index}")

# === Fitur dan Target ===
X = df[["Diameter Buah (cm)", "Berat Buah (gram)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 1. Tampilkan Pohon Keputusan (Teks) ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
# Menggunakan class_names agar hasilnya menampilkan nama spesies, bukan angka
tree_rules = export_text(clf, 
                         feature_names=list(X.columns), 
                         class_names=le.classes_.tolist())
print(tree_rules)
````
