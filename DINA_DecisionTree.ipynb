### 3️⃣ **Ambil contoh threshold split**

### 4️⃣ **Entropy Awal**

Rumus:

$$
Entropy(D) = -p_1 \log_2(p_1) - p_2 \log_2(p_2)
$$

Dengan:

* $p_1 = \frac{8}{15}$
* $p_2 = \frac{7}{15}$

Maka:

$$
Entropy(D) = -\frac{8}{15} \log_2 \frac{8}{15} - \frac{7}{15} \log_2 \frac{7}{15}
$$

Hitung:

* $\frac{8}{15} \approx 0.533$
* $\frac{7}{15} \approx 0.467$

$$
Entropy(D) \approx -(0.533 \cdot \log_2 0.533) - (0.467 \cdot \log_2 0.467)
$$

* $\log_2 0.533 \approx -0.906$
* $\log_2 0.467 \approx -1.096$

Sehingga:

$$
Entropy(D) \approx -(0.533 \cdot -0.906) - (0.467 \cdot -1.096)
$$

$$
Entropy(D) \approx 0.483 + 0.511 \approx 0.994
$$

**Entropy Awal ≈ 0.994**

---

### 5️⃣ **Entropy Setelah Split**

* **Group Left (≤ 7.0 cm):**

  * 8 data, semua *Cucumis sativus*
  * Entropy = 0 (karena pure class)

* **Group Right (> 7.0 cm):**

  * 7 data, semua *Cucumis melo*
  * Entropy = 0 (karena pure class)

---

### 6️⃣ **Information Gain**

$$
IG = Entropy(D) - \left( \frac{8}{15} \cdot 0 + \frac{7}{15} \cdot 0 \right)
$$

Sehingga:

$$
IG = 0.994 - 0 = \boxed{0.994}
$$

---

### 7️⃣ **Gini Index**

#### Gini Awal

$$
Gini(D) = 1 - p_1^2 - p_2^2
$$

$$
Gini(D) = 1 - \left( \frac{8}{15} \right)^2 - \left( \frac{7}{15} \right)^2
$$

$$
Gini(D) = 1 - 0.284 - 0.218 = 0.498
$$

**Gini Awal ≈ 0.498**

#### Gini Setelah Split

* **Group Left (pure class): Gini = 0**
* **Group Right (pure class): Gini = 0**

#### Gini Split:

$$
Gini_{split} = \frac{8}{15} \cdot 0 + \frac{7}{15} \cdot 0 = 0
$$

## Visualisasi Decision Tree

Ini adalah code program untuk mencari decision tree sekaligus rules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Baru ===
data = {
    "Diameter Buah (cm)": [4.2, 4.5, 4.0, 3.8, 4.3, 16, 17, 16.5, 15, 17.5, 4.8, 4.1, 14, 4.3, 15.5],
    "Berat Buah (gram)": [500, 530, 480, 430, 510, 1800, 1900, 1850, 1750, 2000, 550, 500, 1600, 510, 1800],
    "Kelas": [
        "Cucumis sativus", "Cucumis sativus", "Cucumis sativus", "Cucumis sativus", "Cucumis sativus",
        "Cucumis melo", "Cucumis melo", "Cucumis melo", "Cucumis melo", "Cucumis melo",
        "Cucumis sativus", "Cucumis sativus", "Cucumis melo", "Cucumis sativus", "Cucumis melo"
    ]
}

df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # sativus=0, melo=1

# === Fitur dan Target ===
X = df[["Diameter Buah (cm)", "Berat Buah (gram)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 1. Tampilkan Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)
```

## Visualisasi Splitting

```python
# === Uji 2 Data Baru ===
uji_baru = pd.DataFrame({
    'Diameter Buah (cm)': [5.0, 14.5],
    'Berat Buah (gram)': [600, 1700]
})

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (d, b) in enumerate(zip(uji_baru['Diameter Buah (cm)'], uji_baru['Berat Buah (gram)'])):
    print(f"Data {i+1}: Diameter={d}, Berat={b} → Prediksi: {hasil[i]}")

# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step_diameter = 0.1
plot_step_berat = 10

x_min, x_max = X['Diameter Buah (cm)'].min() - 1, X['Diameter Buah (cm)'].max() + 1
y_min, y_max = X['Berat Buah (gram)'].min() - 100, X['Berat Buah (gram)'].max() + 100

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step_diameter),
                     np.arange(y_min, y_max, plot_step_berat))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Diameter Buah (cm)', 'Berat Buah (gram)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Diameter Buah (cm)'], X.iloc[idx]['Berat Buah (gram)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

plt.xlabel("Diameter Buah (cm)")
plt.ylabel("Berat Buah (gram)")
plt.title("Decision Boundary (Diameter vs Berat Buah)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
